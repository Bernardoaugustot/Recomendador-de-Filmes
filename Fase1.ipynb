# ! Extrair os dados do zip ❌
Eles estão todos com o mesmo nome precisam ser extraidos

❌❌❌ PULAR ❌❌❌

```python
from rarfile import RarFile  
import csv

caminho_do_arquivo_rar = 'data/name.basics.tsv.gz'
nome_do_arquivo_csv = "data.tsv"

dados_lidos = []

# Abre o arquivo WinRAR
with RarFile(caminho_do_arquivo_rar, 'r') as arquivo_rar:
    # Verifica se o arquivo CSV está no WinRAR
    if nome_do_arquivo_csv in arquivo_rar.namelist():
        # Extrai o arquivo CSV para uma pasta temporária ou para o diretório atual
        arquivo_rar.extract(nome_do_arquivo_csv)
        
        # Agora podemos ler o arquivo CSV normalmente
        with open(nome_do_arquivo_csv, 'r', newline='', encoding='utf-8') as arquivo_csv:
            leitor_csv = csv.reader(arquivo_csv)
            
            # Armazena os dados lidos em uma lista
            dados_lidos = [linha for linha in leitor_csv]
    else:
        print(f"O arquivo '{nome_do_arquivo_csv}' não foi encontrado no WinRAR.")
```

### Corelacionar filmes por genero.
  Agora vamos começar a trabalhar na recomendação, temos os filmes ordernados por relevancia, vamos então escolher um filme e trazer outros com alta relevancia e generos parecidos.
* tconst è o ID dos filmes.
* Os generos dos filmes estão no formato: Action,Adventure,Sci-Fi	
    por isso teremos que splitar e rodar algumas vezes, fazer com que os filmes que aparecer mais vezes tenham maior prioridade. (talvez usar uma multiplicação pode ajudar.)

```python
def Descobrir_genero(df, filme = 'tt0137523'):

    # !!! Isso só funciona caso tenha apenas 1 genero. 
    # Encontra o índice da linha que contém o filme.
    indice_filme = df.loc[df['tconst'] == filme].index[0]
    # Obtém o valor da coluna "genres" para o filme.
    genero_filme = df.loc[indice_filme, 'genres']

    # Filtrando nosso DataFrame para conter apenas filmes compativeis.
    df_generos = df[df['genres'].str.contains( genero_filme, case=False)]
    df_generos = df_generos.sort_values(by='relevancia', ascending=False)

    return df_generos.head()

Descobrir_genero(df_relevancia)
#df_relevancia.head()
```

Aqui ja temos uma filtragem dos filmes baseados em generos, e relevancia, conseguimos buscar filmes relevantes e ccom generos similates, ainda precisa de polimento mas é um começo.
